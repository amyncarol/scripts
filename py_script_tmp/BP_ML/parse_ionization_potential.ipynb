```python
from bs4 import BeautifulSoup
```

```python
def get_ips(file, number):
    with open(file, 'r') as f:
        html = f.read()
    parsed_html = BeautifulSoup(html, "lxml")
    html_list = parsed_html.body.findAll('td', attrs={'role':'gridcell'})
    ip_dict = {}
    for i in range(number):
        ips = []
        element = html_list[i*33+2].text
        for j in range(3,33):
            ip = html_list[i*33+j].text
            if ip !='':
                ips.append(float(ip))
        ip_dict[element] = ips
    return ip_dict
```

```python
numbers = [19, 20, 20, 20, 20]
ip_dict_all = {}
for i in range(5):
    file = '/Users/yao/Desktop/band offset/'+ str(i+1) + '.htm'
    print(i)
    ip_dict = get_ips(file, numbers[i])
    ip_dict_all.update(ip_dict)
    print(len(ip_dict))
```

```python
for i in ip_dict_all:
    print(i, ip_dict_all[i])
```

```python
import json
with open('/Users/yao/Desktop/band offset/ip.json', 'w') as f:
    json.dump(ip_dict_all, f, sort_keys=True, indent=4)
```
