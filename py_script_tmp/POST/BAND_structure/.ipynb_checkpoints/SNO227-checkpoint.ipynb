```python
import matplotlib.pyplot as plt
from BsDos import *

v = Vasprun("/Users/yao/Google Drive/data/SNO/SNO-4/U7/vasprun.xml")
ylim = [-1, 5]
scissor = 0
fig = plt.figure(figsize=(16,10))
gs = gridspec.GridSpec(1, 4, width_ratios=[1, 1, 1, 1]) 

plt.subplot(gs[0])
draw_total_dos(v, [-40, 40], ylim, scissor)

plt.subplot(gs[1])
draw_pdos(v, 'Nb', [-40, 40], ylim, scissor)

plt.subplot(gs[2])
draw_pdos(v, 'O', [-40, 40], ylim, scissor)

plt.show()
```

```python
matplotlib.rcParams.update({'font.size': 12})
fig = plt.figure(figsize=(16,6))
gs = gridspec.GridSpec(1, 9, width_ratios=[1]*9)
plt.subplot(gs[0])
draw_total_dos(v, [-40, 40], ylim)
plt.legend(prop={'size': 8},loc=3)
for i in range(1, 9):
    plt.subplot(gs[i])
    draw_site_dos(v, i+9, [-5, 5], ylim)
    plt.legend(prop={'size': 8},loc=3)
plt.show()
```
