```python
import numpy as np
from scipy.stats import pearsonr
from scipy.stats import spearmanr
item0 = []
item1 = []
item2 = []
with open('/Users/yao/Desktop/band offset/expanded_offset_pbe_exp.csv','r') as f:
    for line in f:
        cal = float(line.split(',')[1])
        exp = float(line.split(',')[2])
        #if line.split(',')[0] != '' and cal >= 0: #for repeated pairs
        if line.split(',')[0] != '': #for unrepeated pairs 
            item0.append((line.split(',')[0]))
            item1.append(cal)
            item2.append(exp)

cpd = np.array(item0)
CAL = np.array(item1)
EXP = np.array(item2)
```

```python
print(len(CAL))
sum(abs(CAL-EXP) < 1)
cpd[abs(CAL-EXP) > 2]
print(np.mean(abs(CAL-EXP)))
print(np.std(abs(CAL-EXP)))
```

```python
r, dummy = pearsonr(CAL, EXP)
print(dummy)
rho, dummy = spearmanr(CAL, EXP)
print(dummy)
import matplotlib.pyplot as plt
plt.plot(CAL, EXP, 'ro')
plt.xlabel(r'Calculated $\Delta E_v$ (eV)')
plt.ylabel('Measured  $\Delta E_v$ (eV)')
plt.legend()

plt.annotate('r = %f' %r,
            xy=(0.2, 0.8), textcoords='figure fraction')
plt.annotate('rho = %f' %rho,
            xy=(0.2, 0.75), textcoords='figure fraction')
#plt.xlim([-1, 7])
#plt.ylim([-1, 7])
plt.plot(range(-7,7), range(-7,7), '-')

from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(CAL, EXP)
#x = range(-7, 7)
#y = [slope*i+intercept for i in x]
#plt.plot(x, y, '-')
#plt.annotate('y = %f + %f * x ' %(intercept, slope), 
#            xy=(0.2, 0.85), textcoords='figure fraction')

#plt.savefig('/Users/yao/Desktop/band offset/pbebs_pbebs_correlation.pdf', dpi=300, format='pdf')
plt.show()
```
