```python
import numpy as np
import matplotlib.pyplot as plt
import re
from matplotlib import rc
import matplotlib
matplotlib.matplotlib_fname()
```

```python
compounds = np.array(['K','Rb','Cs'])
Eg_A2SnI6_cubic = np.array([0.0673, 0.1534, 0.2883])
Eg_A2PtI6_cubic = np.array([0.4242, 0.5144, 0.6561])
Eg_A2TeI6_cubic = np.array([1.2709, 1.3406, 1.5068])

Eg_A2SnI6_tetra = np.array([0.6576, 0.6714, 0.3674])
Eg_A2PtI6_tetra = np.array([0.9131, 0.7456,  0.6174])
Eg_A2TeI6_tetra = np.array([1.7005, 1.6936, 1.4475])



#Eg_A2SnI6_diff = Eg_A2SnI6_tetra-Eg_A2SnI6_cubic
#Eg_A2PtI6_diff = Eg_A2PtI6_tetra-Eg_A2PtI6_cubic
#Eg_A2TeI6_diff = Eg_A2TeI6_tetra-Eg_A2TeI6_cubic
```

```python
comp = np.arange(len(compounds))
comp2 = comp+4
comp3 = comp+8

comp_tetra = comp+0.3
comp_tetra2 = comp2+0.3
comp_tetra3 = comp3+0.3

plt.figure(figsize = (12, 8))
ax = plt.subplot(2, 2, 1)

plt.bar(comp, Eg_A2SnI6_cubic, width=0.3, align='center',color='r', label = 'cubic')
plt.bar(comp2, Eg_A2PtI6_cubic, width=0.3, align='center', color='r')
plt.bar(comp3, Eg_A2TeI6_cubic, width=0.3, align='center', color='r')
plt.bar(comp_tetra[:2], Eg_A2SnI6_tetra[:2], width=0.3,  align='center', color='b', label = 'tetragonal')
plt.bar(comp_tetra2[:2], Eg_A2PtI6_tetra[:2], width=0.3, align='center', color='b')
plt.bar(comp_tetra3[:2], Eg_A2TeI6_tetra[:2],  width=0.3, align='center', color='b')
compounds_list = np.array([r'$K_2SnI_6$',r'$Rb_2SnI_6$',r'$Cs_2SnI_6$',
                           r'$K_2PtI_6$',r'$Rb_2PtI_6$',r'$Cs_2PtI_6$', 
                           r'$K_2TeI_6$',r'$Rb_2TeI_6$',r'$Cs_2TeI_6$'])

el_list = np.array(['K', 'Rb', 'Cs', 'K', 'Rb', 'Cs', 'K', 'Rb', 'Cs',])
plt.xticks(np.concatenate((comp+0.3,comp2+0.3,comp3+0.3)), el_list, fontsize = 14)
plt.legend(loc='upper center', fontsize = 14)
plt.yticks(fontsize = 20)
ax.set_ylabel('band gap (eV)', fontsize = 20)

ax.text(0.08, -0.067, "A=",fontsize=14, transform=ax.transAxes)
ax.text(0.36, -0.067, "A=",fontsize=14, transform=ax.transAxes)
ax.text(0.65, -0.067, "A=",fontsize=14, transform=ax.transAxes)

ax.text(0.15, -0.135, "B=Sn",fontsize=14, transform=ax.transAxes)
ax.text(0.45, -0.135, "B=Pt",fontsize=14, transform=ax.transAxes)
ax.text(0.73, -0.135, "B=Te",fontsize=14, transform=ax.transAxes)

ax.text(-0.18, 0.98, "a",fontsize=25, transform=ax.transAxes)
ax.text(0.01, 0.93, r'$A_2BI_6$',fontsize=14, transform=ax.transAxes)

#plt.savefig('/Users/yao/Desktop/216-figure/A2BI6-trend.jpg',bbox_inches='tight',dpi=600)
#plt.show()
```
