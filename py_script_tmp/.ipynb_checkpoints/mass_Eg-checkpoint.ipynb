```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import re
from matplotlib.pyplot import figure, show, axes, sci
from matplotlib import cm, colors
from matplotlib.font_manager import FontProperties
from numpy import amin, amax, ravel
import numpy.ma as ma
```

```python
with open('/Users/yao/Google Drive/data/2116/Hull', 'r') as f:
    hull = []
    for line in f:
        hull.append(float(line.split()[4]))
```

```python
with open('/Users/yao/Google Drive/data/2116/Eg_direct_indirect', 'r') as f:
    Eg = []
    Eg_direct = []
    compounds = []
    for line in f:
        compounds.append([line.split()[0], line.split()[1], line.split()[2], line.split()[3]])
        Eg.append(float(line.split()[4]))
        Eg_direct.append(float(line.split()[5]))
        
```

```python
with open('/Users/yao/Google Drive/data/2116/mass', 'r') as f:
    mass_e = []
    mass_h = []
    for line in f:
        mass_e.append(float(line.split()[4]))
        mass_h.append(float(line.split()[5]))
```

```python
j = 0
for i in np.arange(len(hull)):
    if Eg[i]<0.5:
        print '{0} {1} {2} {3} {4} {5}'.format(compounds[i][0],compounds[i][1], compounds[i][2],compounds[i][3],Eg[i],mass_e[i])
        j = j+1
print j
```

```python
masked_Eg = ma.masked_where(np.asarray(hull)>0.03, np.asarray(Eg))
masked_Eg_direct = ma.masked_where(np.asarray(hull)>0.03, np.asarray(Eg_direct))
masked_mass_e = ma.masked_where(np.asarray(hull)>0.03, np.asarray(mass_e))
masked_mass_h = ma.masked_where(np.asarray(hull)>0.03, np.asarray(mass_h))
```

```python
fig, ax = plt.subplots()
ax.plot(masked_Eg, masked_mass_e, 'o', label = 'electron', color = 'blue', 
        markeredgecolor = 'none', markersize = 5)
ax.plot(masked_Eg, masked_mass_h, 'o', label = 'hole', color = 'red', 
        markeredgecolor = 'none', markersize = 5)
plt.xlabel('band gap(eV/atom)')
plt.ylabel('effective mass')
plt.axis([0, 1.5, 0, 0.8])
plt.legend()
show()
```
