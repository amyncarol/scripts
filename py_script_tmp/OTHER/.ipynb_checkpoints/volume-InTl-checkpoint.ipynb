```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
import pandas as pd

def get_a_energy_eg(file):
    a = []
    energy = []
    eg = []
    with open(file,'r') as f:
        for line in f:
            l = line.split()
            a.append(float(l[1]))
            energy.append(float(l[2]))
            eg.append(float(l[4]))
    a = pd.DataFrame({'strain': a, 'energy': energy, 'gap':eg})
    b = a.sort(columns='strain')
    return np.array(b['strain']), np.array(b['energy']), np.array(b['gap'])

def get_info5(file5):
    a, energy, dummy = get_a_energy_eg(file5)
    volume = a**3*1e-30  ##in SI units
    energy_si = energy*1.60217662e-19
    p = -np.gradient(energy_si)/np.gradient(volume)/1e9
    return volume/1e-30, p

def get_info10(file10):
    a, energy, dummy = get_a_energy_eg(file10)
    volume = (sqrt(2)*a)**3/4*1e-30  ##in SI units
    energy_si = energy*1.60217662e-19
    p = -np.gradient(energy_si)/np.gradient(volume)/1e9
    return volume/1e-30/2, p

def get_transition_p(file10):
    dummy, dummy, gap = get_a_energy_eg(file10)
    dummy, p = get_info10(file10)
    index = sum(gap>0.01)
    if (sum(gap>0.01)==length(gap)):
        return None
    else:
        return p[-index]
    
def draw_strain(file5, file10):
    a, energy, eg = get_a_energy_eg(file5)
    a_2, energy_2, eg_2 = get_a_energy_eg(file10)

    fig, ax1 = plt.subplots()
    l1 = ax1.plot(a**3, energy/5,'o-',label='perovskite - energy')
    l2 = ax1.plot((sqrt(2)*a_2)**3/8, energy_2/10, 'ro-', label='double perovkite - energy')
    ax1.set_xlabel('volume of the 5-atom cell(angstrom^3)')
    ax1.set_ylabel('energy per atom(eV)')

    ax2 = ax1.twinx()
    l3 = ax2.plot(a**3, eg,'^--',label='perovskite - band gap')
    l4 = ax2.plot((sqrt(2)*a_2)**3/8, eg_2, 'r^--', label='double perovkite - band gap')
    ax2.set_ylabel('Band gap(eV)')

    ls = l1+l2+l3+l4
    labs = [l.get_label() for l in ls]
    ax2.legend(ls, labs, loc="upper center")
    return ax1, ax2
```

```python
A = ['Rb','Cs']
B = ['In','Tl']
X = ['F','Cl','Br','I']

for x in X:
    for b in B:
        for a in A:
            try:
                c5 = a+'1'+b+'1'+x+'3'
                c10 = a+'2'+b+'1'+b+'1'+x+'6'
                ax1, ax2 = draw_strain('/Users/yao/Google Drive/data/2116/InTl/5atom/'+c5+'/Eg_result',\
                '/Users/yao/Google Drive/data/2116/InTl/10atom/'+c10+'/Eg_result')
                ax1.text(0.6, 0.3, c5, transform=ax1.transAxes)
                #plt.savefig('/Users/yao/Google Drive/data/2116/2116 figure/InTl/'+c5+'.pdf', dpi=300, bbox_inches='tight')
                plt.show()
                
                print("for cubic structure")
                volume1, p1 = get_info5('/Users/yao/Google Drive/data/2116/InTl/5atom/'+c5+'/Eg_result')
                print("the volume(5 atom):")
                print(volume1)
                print("corresponding pressure in GPa:")
                print(p1)
                
                print("for charge ordered structure")
                volume2, p2 = get_info10('/Users/yao/Google Drive/data/2116/InTl/10atom/'+c10+'/Eg_result')
                print("the volume(5 atom):")
                print(volume2)
                print("corresponding pressure in GPa:")
                print(p2)
                print(get_transition_p('/Users/yao/Google Drive/data/2116/InTl/10atom/'+c10+'/Eg_result'))
                
            except:
                print('')
```
