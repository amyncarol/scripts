```python
from pymatgen.core.periodic_table import Specie
from pymatgen.matproj.rest import MPRester
from pymatgen.phasediagram.pdmaker import PhaseDiagram
from pymatgen.phasediagram.plotter import PDPlotter
from pymatgen.core.periodic_table import Element
import pymatgen.core.periodic_table as pt
import math
import numpy as np
```

```python
all_elements = pt.all_symbols()
e_set_common = set([])
list_radiation = ['Tc', 'Pm', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Uut', 'Fl', 'Uup', 'Lv', 'Uuo']
for item in all_elements:
    e = Element(item) 
    if (3 in e.common_oxidation_states) & (e.symbol not in list_radiation):
        e_set_common.add(e)
        print "%s: %s" %(e.symbol, e.common_oxidation_states) 
```

```python
listC = []
for item in e_set_common:
    listC.append(item.symbol)
```

```python
listC.remove('Cl')
listC.remove('I')
listC.remove('Cr')
print listC
```

```python
listA = ['K','Rb','Cs']
listB = ['Ag','Cu','Li','Na','K']
listD = ['F','Cl','Br','I']
```

```python
listA_radius = {}
for item in listA:
    listA_radius[item]=Specie(item,1).ionic_radius
print listA_radius   

listB_radius = {}
for item in listB:
    listB_radius[item]=Specie(item,1).ionic_radius
print listB_radius   

listC_radius = {}
for item in listC:
    listC_radius[item]=Specie(item,3).ionic_radius
print listC_radius   

listD_radius = {}
for item in listD:
    listD_radius[item]=Specie(item,-1).ionic_radius
print listD_radius   
```

```python
tolerance_dict = {}
tolerance_list = []
for A in listA:
    for B in listB:
        for C in listC:
            for D in listD:
                t = (listA_radius[A]+listD_radius[D])/((listB_radius[B]+listC_radius[C])/2+listD_radius[D])/math.sqrt(2)
                tolerance_dict[A+B+C+D]=t
                tolerance_list.append(t)
tolerance_data = np.array(tolerance_list)
hist,bins=np.histogram(tolerance_data,bins=[0.7,0.9,1.0,1.2])
print(hist)
print tolerance_dict
```

```python
pauling_dict = {}
pauling_list = []
for B in listB:
    for C in listC:
        for D in listD:
            p = (listB_radius[B]+listC_radius[C])/2/listD_radius[D]
            pauling_dict[B+C+D]=p
            pauling_list.append(p)
pauling_data = np.array(pauling_list)
hist2,bins2=np.histogram(pauling_data,bins=[0,0.414,0.732,1.0])
print(hist2) 
```
