```matlab
errors = zeros(20,1);
for m = 5:5:100;
    b = randn(m, 1);
    A = zeros(m);
    for i=1:m
        A(i,i) = 1;
        A(i, m) = 1;
        for j = 1:(i-1)
            A(i, j) = -1;
        end
    end
    x = A\b;
    errors(m/5,1) = norm(A*x - b)/norm(A);
    norm(A)
end
format shortE
errors
```
