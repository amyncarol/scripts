```python
#This is to use the 216paper env and pymatgen4.5.7 with python3.5, I really hate hwo pymatgen updated!!!!!
#and matplotlib 2.0.1
from pymatgen.io.vasp.outputs import Vasprun
import matplotlib.pyplot as plt
import matplotlib
import os
import numpy as np
from matplotlib import gridspec
matplotlib.rcParams.update({'font.size': 20})
```

```python
def get_k_dist(kpoints):
    dist = np.zeros(len(kpoints))
    for i in range(len(kpoints)-1):
        dist[i+1]=dist[i]+np.linalg.norm(kpoints[i]-kpoints[i+1])
    return dist
    
```

```python
def get_special_k(kpoints):
    k_index = [0]
    for i in range(len(kpoints)-1):
        if np.linalg.norm(kpoints[i]-kpoints[i+1]) < 1e-8:
            k_index.append(i)
    k_index.append(len(kpoints)-1)
    return get_k_dist(kpoints)[k_index]
```

```python
def get_k_label(kpoints, labels):
    k_index = [0]
    for i in range(len(kpoints)-1):
        if np.linalg.norm(kpoints[i]-kpoints[i+1]) < 1e-8:
            k_index.append(i)
    k_index.append(len(kpoints)-1)
    return labels[k_index]
```

```python
##This is just for bands with up spin(in a spin-polarized calculation for non-magnetic materials)
def draw_bands(bs, ylim):
    kpoints = [i.cart_coords for i in bs.kpoints]
    kpoints_dist = get_k_dist(kpoints)
    label = get_k_label(kpoints, np.array([i.label for i in bs.kpoints]))
    bands = [bs.bands[i] for i in bs.bands]
    bands_up = np.array(bands[0]) - bs.get_vbm()['energy']
    for i in range(len(bands_up)):
        plt.plot(kpoints_dist, bands_up[i],'b')
    plt.plot([0, kpoints_dist[-1]], [0, 0], 'k--')
    for i in get_special_k(kpoints):
        plt.plot([i, i], ylim, 'k')
    plt.xlim([0, kpoints_dist[-1]])
    plt.ylim(ylim)
    label_loc = get_special_k(kpoints)
    label[label=='\\Gamma'] = r'$\Gamma$'
    plt.xticks(label_loc, label)
    plt.ylabel('Energy (eV)')
    
```

```python
## This is just for SOC bands ksplit calculations
def draw_bands_ksplit(kpoints_obj, bands, vbm, ylim):
    kpoints = [i.cart_coords for i in kpoints_obj]
    kpoints_dist = get_k_dist(kpoints)
    label = get_k_label(kpoints, np.array([i.label for i in kpoints_obj]))
    bands = bands - vbm
    for i in range(len(bands)):
        plt.plot(kpoints_dist, bands[i],'b')
    plt.plot([0, kpoints_dist[-1]], [0, 0], 'k--')
    for i in get_special_k(kpoints):
        plt.plot([i, i], ylim, 'k')
    plt.xlim([0, kpoints_dist[-1]])
    plt.ylim(ylim)
    label_loc = get_special_k(kpoints)
    label[label=='\\Gamma'] = r'$\Gamma$'
    plt.xticks(label_loc, label)
    plt.ylabel('Energy (eV)')
```

```python
#draw total DOS only
def draw_total_dos(v, xlim, ylim):
    vbm = v.eigenvalue_band_properties[2]
    comp_dos = v.complete_dos
    elements = [i for i in comp_dos.get_element_dos()]
    for element in elements:
        el_dos = comp_dos.get_element_dos()[element]
        plt.plot(el_dos.get_densities(), el_dos.energies-vbm, label = element)

    tdos = v.tdos
    plt.plot(tdos.get_densities(), tdos.energies-vbm, 'k', label = 'Total')

    plt.legend()
    plt.ylim(ylim)
    plt.xlim(xlim)
    plt.plot(xlim, [0, 0], 'k--')
    plt.xlabel('DOS (states/eV)')
    plt.yticks([])

```

```python
#draw pDOS only
def draw_pdos(v, element_no, xlim, ylim):
    vbm = v.eigenvalue_band_properties[2]
    comp_dos = v.complete_dos
    elements = [i for i in comp_dos.get_element_dos()]
    element = elements[element_no]
    pdos = comp_dos.get_element_spd_dos(element)

    orbitals = [i for i in pdos]
    for o in orbitals:
        p_dos = pdos[o]
        plt.plot(p_dos.get_densities(), p_dos.energies-vbm, label = element.symbol + '-' + o.name)

    plt.legend()
    plt.ylim(ylim)
    plt.xlim(xlim)
    plt.plot(xlim, [0, 0], 'k--')
    plt.xlabel('DOS (states/eV)')
    plt.yticks([])

```

```python
fig = plt.figure(figsize=(16,10))
gs = gridspec.GridSpec(1, 4, width_ratios=[2, 1, 1, 1]) 

ax = plt.subplot(gs[0])

ax.text(-0.2, 1, 'a', fontsize=40, transform=ax.transAxes)
v = Vasprun("/Users/yao/Google Drive/data/216/cubic-hse/Cs2Te1I6/hse-bands2/vasprun.xml")
bs = v.get_band_structure(line_mode=True)
draw_bands(bs, [-1, 4])

plt.subplot(gs[1])
v = Vasprun("/Users/yao/Google Drive/data/216/cubic-hse/Cs2Te1I6/vasprun.xml")
draw_total_dos(v, [0, 40], [-1, 4])

plt.subplot(gs[2])
draw_pdos(v, 0, [0, 20], [-1, 4])

plt.subplot(gs[3])
draw_pdos(v, 1, [0, 20], [-1, 4])



plt.savefig('/Users/yao/Desktop/216-figure/BS_DOS/Cs2Te1I6_HSE.pdf',dpi=300, bbox_inches='tight')
plt.show()
```

```python
def add_delete_nbands(nbands, bands):
    nbands_2 = bands.shape[0]
    if nbands_2 < nbands:
        bands = np.concatenate((bands, np.ones((nbands-nbands_2, bands.shape[1]))*100))
    if nbands_2 > nbands:
        bands = bands[nbands, :]
    return bands
```

```python
fig = plt.figure(figsize=(16,10))
gs = gridspec.GridSpec(1, 4, width_ratios=[2, 1, 1, 1]) 

ax = plt.subplot(gs[0])
ax.text(-0.2, 1, 'b', fontsize=40, transform=ax.transAxes)

for i in range(1,6):
    v = Vasprun("/Users/yao/Google Drive/data/216/cubic-hse/Cs2Te1I6/hse_soc_bands/k"+str(i)+"/vasprun.xml")
    bs = v.get_band_structure(line_mode=True)
    if i == 1:
        kpoints = bs.kpoints
        bands = [bs.bands[i] for i in bs.bands][0]
        nbands = bands.shape[0]
        vbm = bs.get_vbm()['energy']
    if i > 1:
        kpoints = np.concatenate((kpoints, bs.kpoints))
        bands = np.concatenate((bands, add_delete_nbands(nbands, [bs.bands[i] for i in bs.bands][0])), axis=1)
        if bs.get_vbm()['energy'] > vbm:
            vbm = bs.get_vbm()['energy']
draw_bands_ksplit(kpoints, bands, vbm, [-1, 4])

plt.subplot(gs[1])
v = Vasprun("/Users/yao/Google Drive/data/216/cubic-hse/Cs2Te1I6/SOC/vasprun.xml")
draw_total_dos(v, [0, 40], [-1, 4])

plt.subplot(gs[2])
draw_pdos(v, 0, [0, 20], [-1, 4])

plt.subplot(gs[3])
draw_pdos(v, 1, [0, 20], [-1, 4])

plt.savefig('/Users/yao/Desktop/216-figure/BS_DOS/Cs2Te1I6_HSE_SOC.pdf',dpi=300, bbox_inches='tight')
plt.show()
```
