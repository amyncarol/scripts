```python
import matplotlib.pyplot as plt
from BsDos import *

v = Vasprun("/Users/yao/Google Drive/data/SNO/SNO-4/U7/vasprun.xml")
ylim = [-1, 5]
scissor = 0
fig = plt.figure(figsize=(16,10))
gs = gridspec.GridSpec(1, 4, width_ratios=[1, 1, 1, 1]) 

plt.subplot(gs[0])
draw_total_dos(v, [-40, 40], ylim, scissor)

plt.subplot(gs[1])
draw_pdos(v, 'Nb', [-40, 40], ylim, scissor)

plt.subplot(gs[2])
draw_pdos(v, 'O', [-40, 40], ylim, scissor)
plt.savefig('/Users/yao/Google Drive/data/SNO/SNO-4/U7/DOS1.pdf')
plt.show()
```

```python
matplotlib.rcParams.update({'font.size': 12})
fig = plt.figure(figsize=(9,6))
gs = gridspec.GridSpec(1, 3, width_ratios=[1]*3)
plt.subplot(gs[0])
draw_total_dos(v, [-40, 40], ylim)
plt.legend(prop={'size': 8},loc='center')
for i in range(1, 3):
    plt.subplot(gs[i])
    draw_site_dos(v, (i-1)*4+9, [-5, 5], ylim)
    plt.legend(prop={'size': 8},loc='center')
plt.savefig('/Users/yao/Google Drive/data/SNO/SNO-4/U7/DOS2.pdf')
plt.show()
print('Nb 9: close to oxygen layer')
```

```python
matplotlib.rcParams.update({'font.size': 12})
fig = plt.figure(figsize=(14,5))
gs = gridspec.GridSpec(1, 8, width_ratios=[1]*8)
plt.subplot(gs[0])
draw_total_dos(v, [-40, 40], ylim)
plt.legend(prop={'size': 8}, loc='center')
for i in range(1,8):
    plt.subplot(gs[i])
    draw_site_dos(v, (i-1)*4+17, [-1, 1], ylim)
    plt.legend(prop={'size': 8},loc='center')
plt.savefig('/Users/yao/Google Drive/data/SNO/SNO-4/U7/DOS3.pdf')
plt.show()
print('close to oxygen layer-->1, far from oxygen layer-->5'+'\n'+ 
    '17:5, 21:3, 25: 2, 29: 1, 33:1, 37: 3, 41: 4')
```
