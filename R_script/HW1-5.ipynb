```r
quakes <- read.table(file = "/Users/yao/Desktop/stat201b/HW/HW1/fijiquakes.dat", header=TRUE)
```

```r
min = range(quakes$mag)[1]-0.1
max = range(quakes$mag)[2]
n = length(quakes$mag)
```

```r
x = seq(min, max, length = 100)
Fn = rep(0, 100)
for (i in 1:100){
    Fn[i] = sum(quakes$mag <= x[i])/1000
}
alpha = 0.05
epsilon = sqrt(log(2/alpha)/(2*n))
Fn_lower = rep(0, 100)
Fn_upper = rep(0, 100)
for (i in 1:100){
    Fn_lower[i] = max(Fn[i]-epsilon, 0)
    Fn_upper[i] = min(Fn[i]+epsilon, 1)
}
plot(x, Fn, type='l')
lines(x, Fn_lower, type='l', lty=3)
lines(x, Fn_upper, type='l', lty=3)
```

```r
apply(outer(quakes$mag, x, "<="),2,mean)
```
