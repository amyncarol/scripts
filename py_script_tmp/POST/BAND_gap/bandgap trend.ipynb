```python
import numpy as np
import matplotlib.pyplot as plt
import re
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
rc('font', **font)
rc('xtick', labelsize=22)
```

```python
# delta-sol_without_corr

comp = np.array([0, 1.0/3.0, 2.0/3.0, 1])
compounds = np.array(['CsSnI3 orth', 'CsSnI2Br orth', 'CsSnIBr2', 'CsSnBr3'])
bandgap_norelax = np.array([2.04, 2.15, 2.15, 2.28])
bandgap_relax = np.array([1.92, 1.99, 2.05, 2.28])
image_corr = np.array([-0.65, -0.74, -0.86, -1.03])
bandgap_norelax_corr = bandgap_norelax + image_corr
bandgap_relax_corr = bandgap_relax + image_corr
atomic_volume = np.array([48.595, 46.4835, 43.8575, 40.868])

bandgap_exp = np.array([1.27, 1.37, 1.65, 1.75])
lattice_exp = np.array([6.16, 6.10, 5.92, 5.80])
atomic_volume_exp = np.array([46.75, 45.40, 41.49, 39.02])

fig, ax = plt.subplots()
ax.plot(atomic_volume, bandgap_norelax,'-o', label = 'delta-sol', linewidth = 2, markersize = 10)
for x, y, z in zip(atomic_volume, bandgap_relax, compounds):
    ax.text(x, y, '%s' % z)
#ax.plot(atomic_volume, bandgap_norelax,'-o', label = 'norelax')
#for x, y, z in zip(atomic_volume, bandgap_norelax, compounds):
#    ax.text(x, y, '%s' % z)
ax.plot(atomic_volume_exp, bandgap_exp,'-o', label = 'experiment', linewidth = 2, markersize = 10)
for x, y, z in zip(atomic_volume_exp, bandgap_exp, compounds):
    ax.text(x, y, '%s' % z) 
    
#ax0.set_xlabel('Br/(Br+I)')
ax.set_ylabel('bandgap (eV)')
ax.set_xlabel(r'atomic volume($\AA^3$)')
ax.legend(loc='lower left')
plt.show()

```

```python
# figure4
comp = np.array([0, 1.0/3.0, 2.0/3.0, 1])
compounds = np.array(['CsSnI3', 'CsSnI2Br', 'CsSnIBr2', 'CsSnBr3'])
E_cubic = np.array([-2.809, -2.923, -3.043, -3.169])
E_orth = np.array([-2.818, -2.929, -3.048, -3.173])
E_diff = E_cubic - E_orth

fig, ax = plt.subplots()
ax.plot(comp, E_diff,'-o', label ='E_cubic-E_orthorhombic')
for x, y, z in zip(comp, E_diff, compounds):
    ax.text(x+0.02, y, '%s' % z)

#ax0.set_xlabel('Br/(Br+I)')
ax.set_ylabel('energy difference (eV/atom)')
ax.set_xlabel('Br/(Br+I)')
ax.legend(loc='upper right')
plt.show()
```
