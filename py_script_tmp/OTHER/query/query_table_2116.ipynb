```python
def get_minus1():
    B_list = []
    list_radiation = ['Tc', 'Pm', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Uut', 'Fl', 'Uup', 'Lv', 'Uuo']
    for z in range(1, 104):
        B = Element.from_Z(z)
        if -1 in B.common_oxidation_states and B.symbol not in list_radiation and B.symbol not in ['H']:
            B_list.append(B.symbol) 
    return B_list

def get_plus1():
    B_list = []
    list_radiation = ['Tc', 'Pm', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Uut', 'Fl', 'Uup', 'Lv', 'Uuo']
    for z in range(1, 104):
        B = Element.from_Z(z)
        if 1 in B.common_oxidation_states and B.symbol not in list_radiation and B.symbol not in ['H', 'I', 'Cl', 'Br']:
            B_list.append(B.symbol) 
    B_list += ['Cu', 'Au', 'In']
    return B_list

def get_plus2():
    B_list = []
    list_radiation = ['Tc', 'Pm', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Uut', 'Fl', 'Uup', 'Lv', 'Uuo']
    for z in range(1, 104):
        B = Element.from_Z(z)
        if 2 in B.common_oxidation_states and B.symbol not in list_radiation:
            B_list.append(B.symbol) 
    return B_list

def get_plus3():
    B_list = []
    list_radiation = ['Tc', 'Pm', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Uut', 'Fl', 'Uup', 'Lv', 'Uuo']
    for z in range(1, 104):
        B = Element.from_Z(z)
        if 3 in B.common_oxidation_states and B.symbol not in list_radiation and B.symbol not in ['H', 'I', 'Cl', 'Br']:
            B_list.append(B.symbol) 
    B_list += ['Ta','Mo']
    return B_list

def get_plus4():
    B_list = []
    list_radiation = ['Tc', 'Pm', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Uut', 'Fl', 'Uup', 'Lv', 'Uuo']
    for z in range(1, 104):
        B = Element.from_Z(z)
        if 4 in B.common_oxidation_states and B.symbol not in list_radiation:
            B_list.append(B.symbol)
    B_list.append(u'Ni') 
    return B_list
```

```python
print(get_plus1())
print(get_plus3())
print(get_minus1())
with MPRester(os.environ['MAPI_KEY']) as m:
    results = []
    i = 0
    for A in get_plus1():
        for B in get_plus1():
            for Bp in get_plus3():
                for X in get_minus1():
                    #print(A+"2"+B+Bp+X+"6")
                    result = m.query(criteria={"pretty_formula": A+"2"+B+Bp+X+"6"}, properties=["pretty_formula", 
                                                                                         "spacegroup.symbol",
                                                                                         "nsites",
                                                                                         "icsd_id",
                                                                                         "e_above_hull",
                                                                                         "band_gap",
                                                                                         "task_id"
                                                                                        ])
                    
                    if len(result)!=0:
                        for j in result:
                            results.append(j)
                    i += 1
                    if i%100 == 0: print(i)
```

```python
import collections

data = collections.defaultdict(list)
for e in results:
    data["Materials ID"].append(e["task_id"])
    data["Icsd ID"].append(str(e["icsd_id"]))
    data["Formula"].append(e["pretty_formula"]) 
    data["Spacegroup"].append(e["spacegroup.symbol"]) 
    data["Nsites"].append(e["nsites"]) 
    data["Ehull"].append(e["e_above_hull"]) 
    data["Bandgap"].append(e["band_gap"]) 

from pandas import DataFrame
import pandas as pd
def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    
df = DataFrame(data, columns=[
        "Materials ID", "Icsd ID", "Formula", "Spacegroup", "Nsites", "Ehull", "Bandgap"])

df
```

```python
def get_tolerance_factor(A, B, Bp, X):
    import math
    #Shannon ionic radii with oxidation state and coordination  (and some with Zachariasen's relationship)
    A_radius = {'Li':1.13, 'Na':1.39, 'K':1.64, 'Rb':1.72, 'Cs':1.88, 'Tl':1.7}   #12 coodination

    B_radius = {'Li':0.76, 'Na':1.02, 'K':1.38, 'Rb':1.52, 'Cs':1.67, 'Cu':0.77,'Ag':1.15,'Au':1.37, 'Tl':1.5}  #6 coodination

    Bp_radius = {'Bi':1.03,'Sb':0.76,'In':0.8,'Au':0.85,'Ta':0.72,'Mo':0.69,'Sc':0.745,'Y':0.9,'La':1.032,\
                 'Ce':1.01,'Pr':0.99,'Nd':0.983,'Sm':0.958,'Eu':0.947,'Gd':0.938,'Tb':0.923,'Dy':0.912,\
                 'Er':0.89,'Tm':0.88,'Lu':0.861, 'Cr':0.615, 'Fe':0.645, 'Ga':0.62, 'Ho':0.901, 'Yb':0.868,\
                 'Al':0.535, 'Tl':0.885, 'Rh':0.665 }   #6 coodination #Fe:high spin

    X_radius = {'F':1.33,'Cl':1.81,'Br':1.96,'I':2.2}   #2+4=6 coodination
    try:
        return (A_radius[A]+X_radius[X])/((B_radius[B]+Bp_radius[Bp])/2+X_radius[X])/math.sqrt(2)
    except:
        return "Failed"
    
def get_oct_factor(A, B, Bp, X):
    #Shannon ionic radii with oxidation state and coordination  (and some with Zachariasen's relationship)
    A_radius = {'Li':1.13, 'Na':1.39, 'K':1.64, 'Rb':1.72, 'Cs':1.88, 'Tl':1.7}   #12 coodination

    B_radius = {'Li':0.76, 'Na':1.02, 'K':1.38, 'Rb':1.52, 'Cs':1.67, 'Cu':0.77,'Ag':1.15,'Au':1.37, 'Tl':1.5}  #6 coodination

    Bp_radius = {'Bi':1.03,'Sb':0.76,'In':0.8,'Au':0.85,'Ta':0.72,'Mo':0.69,'Sc':0.745,'Y':0.9,'La':1.032,\
                 'Ce':1.01,'Pr':0.99,'Nd':0.983,'Sm':0.958,'Eu':0.947,'Gd':0.938,'Tb':0.923,'Dy':0.912,\
                 'Er':0.89,'Tm':0.88,'Lu':0.861, 'Cr':0.615, 'Fe':0.645, 'Ga':0.62, 'Ho':0.901, 'Yb':0.868,\
                 'Al':0.535, 'Tl':0.885, 'Rh':0.665 }   #6 coodination #Fe:high spin

    X_radius = {'F':1.33,'Cl':1.81,'Br':1.96,'I':2.2}   #2+4=6 coodination

    try:
        return (B_radius[B]+Bp_radius[Bp])/2/X_radius[X]
    except:
        return 'Failed'
    
def get_elements(compound):
    import re
    m = re.match('([A-Z][a-z]?)(\d)([A-Z][a-z]?)([A-Z][a-z]?)([A-Z][a-z]?)(\d)', compound)
    if m:
        return m.group(1), m.group(3), m.group(4), m.group(5)
```

```python
df['Tolerance_factor'] = df['Formula'].map(lambda x: get_tolerance_factor(\
                get_elements(x)[0], get_elements(x)[1], get_elements(x)[2], get_elements(x)[3]))
df['Oct_factor'] = df['Formula'].map(lambda x: get_oct_factor(\
                get_elements(x)[0], get_elements(x)[1], get_elements(x)[2], get_elements(x)[3]))

```
