```python
import pandas as pd
from pymatgen.core.periodic_table import Element 
import numpy as np
```

```python
df = pd.DataFrame(columns=['A','B1','B2','X','AB1B2X'], index=np.arange(36))

n = 0
for A in ['K','Rb','Cs']:
    for B1 in ['In','Tl']:
        for B2 in ['In','Tl']:
            for X in ['Cl','Br','I']:
                df.iloc[n, 0] = A
                df.iloc[n, 1] = B1
                df.iloc[n, 2] = B2
                df.iloc[n, 3] = X
                df.iloc[n, 4] = A + B1 + B2 + X
                n = n+1
```

```python
def get_finished():
    A=['K','Rb','Cs']
    B=['Ag','Cu','K','Na','Li']
    C=['Bi','Sb','Ce','Pr','Nd','Ta','Mo','Y']
    D=['F','Cl','Br','I']
    list1 = []
    for a in A:
        for b in B:
            for c in C:
                for d in D:
                    list1.append(a+b+c+d)
                    
                    
    with open('/Users/yao/Google Drive/data/2116/screen_list_Cl','r') as f:
        for line in f:
            list1.append(line.split(' ')[0]+line.split(' ')[1]+line.split(' ')[2]+line.split(' ')[3].replace('\n', ''))
            
    with open('/Users/yao/Google Drive/data/2116/screen_list_FBrI','r') as f:
        for line in f:
            list1.append(line.split(' ')[0]+line.split(' ')[1]+line.split(' ')[2]+line.split(' ')[3].replace('\n', ''))
            
    with open('/Users/yao/Google Drive/data/2116/screen_list_remaining_wo_transition_metal','r') as f:
        for line in f:
            list1.append(line.split(' ')[0]+line.split(' ')[1]+line.split(' ')[2]+line.split(' ')[3].replace('\n', ''))
            
    return set(list1)
```

```python
compute_df = df[~df['AB1B2X'].isin(get_finished())]
```

```python
compute_df.to_csv('/Users/yao/Google Drive/data/2116/screen_list_InTl', \
                  sep=' ',columns=['A','B1','B2','X'],  header=False, index=False)
```
