```python
data = open('/Users/yao/Desktop/Cl-compounds-tolerance','r')
data_output = open ('/Users/yao/Desktop/screen_list_Cl','w')
```

```python
A=['K','Rb','Cs']
B=['Ag','Cu','K','Na','Li']
C=['Bi','Sb','Ce','Pr','Nd','Ta','Mo','Y']
D=['F','Cl','Br','I']
```

```python
compounds = []
compounds_exclude = []
lines = data.readlines()
for line in lines:
    match = re.match(r"([a-z]+)([\t]+)([a-z]+)([\t]+)([a-z]+)([\t]+)([a-z]+)", line, re.I)
    if match:
        items = match.groups()
        if not ((items[0] in A) & (items[2] in B) & (items[4] in C) & (items[6] in D)):
            compounds.append([items[0], items[2], items[4], items[6]])
            print >> data_output, items[0], items[2], items[4], items[6]
        else:
            compounds_exclude.append([items[0], items[2], items[4], items[6]])
```
