```python
from pymatgen.io.vasp.outputs import Vasprun, Outcar
import os
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
basedir = "/Users/yao/Google Drive/data/216/vasprun-files"
element_list = {'Ti':0, 'Zr':1, 'Hf':2, "V":3, "Nb":4, "Ta":5, "Cr":6, "Mo":7, "W":8, "Mn":9, "Tc":10, "Re":11, "Fe":12,
               "Ru":13, "Os":14, "Co":15, "Rh":16, "Ir":17, "Ni":18, "Pd":19, "Pt":20, "Cu":21, "Ag":22, "Hg":23, 
               "Si":24, "Ge":25, "Sn":26, "Pb":27, "Se":28, "Te":29, "Ce":30, "Pr":31, "Tb":32}
nelect_list = {'Ti':10, 'Zr':12, 'Hf':10, "V":11, "Nb":11, "Ta":11, "Cr":12, "Mo":12, "W":12, "Mn":13, "Tc":13, "Re":13, 
               "Fe":14,"Ru":14, "Os":14, "Co":9, "Rh":15, "Ir":9, "Ni":16, "Pd":10, "Pt":10, "Cu":17, "Ag":11, "Hg":12, 
               "Si":4, "Ge":14, "Sn":14, "Pb":14, "Se":6, "Te":6, "Ce":12, "Pr":11, "Tb":9}
j = 0
items = os.listdir(basedir)
data = np.zeros((len(items), 3))
for item in items:
    outcar = Outcar(os.path.join(basedir, item, "OUTCAR"))
    tot_charge = outcar.charge[2]["tot"]  
    elements = re.split("[0-9]+", item)
    data[j][0] = element_list[elements[1]]
    data[j][1] = tot_charge  
    oxidation_state = nelect_list[elements[1]]-tot_charge
    data[j][2] = oxidation_state
    j = j+1
data
```

```python
plt.plot(data[:, 0], data[:, 1], 'o')

list_number = []
list_element = []
for i in element_list:
    #print i, element_list[i]
    list_number.append(element_list[i]) 
    list_element.append(i)
plt.xticks(list_number, list_element)

plt.xlim(0, len(element_list)-1)
plt.ylim(0, 20)

for i in range(0, len(list_number)):
    plt.axvline(x=i, ymin=0, ymax = 4, linewidth=1, linestyle='--', color='k')

for element in nelect_list:
    nelect = nelect_list[element] 
    posi = element_list[element]
    scale = len(element_list)-1
    plt.axhline(y = nelect, xmin = (posi-0.5)/scale, xmax = (posi+0.5)/scale, linewidth = 2, color='r')

plt.show()
```
