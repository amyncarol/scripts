```python
import matplotlib.pyplot as plt
import re
import numpy as np

e_mf = []
e_cor = []
with open('/Users/yao/Google Drive/data/BGW-test/Si/val','r') as f:
    for line in f:
        e_mf.append(float(line.split()[0]))
        e_cor.append(float(line.split()[1]))
        
with open('/Users/yao/Google Drive/data/BGW-test/Si/cond','r') as f:
    for line in f:
        e_mf.append(float(line.split()[0]))
        e_cor.append(float(line.split()[1]))

plt.plot(e_mf,e_cor,'o')
plt.plot([6.28493,6.28493],[-1,1],'r')
plt.plot([6.977091,6.977091],[-1,1],'r')

a_val,b_val = np.polyfit(e_mf[25:32], e_cor[25:32], deg=1)
print(a_val,b_val)
x = np.arange(0,15,0.1)
plt.plot(x, a_val*x+b_val)
a_cond,b_cond = np.polyfit(e_mf[32:40], e_cor[32:40], deg=1)
print(a_cond,b_cond)
plt.plot(x, a_cond*x+b_cond)
plt.show()

```
